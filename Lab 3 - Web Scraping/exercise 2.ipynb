```python
import requests
from bs4 import BeautifulSoup
import csv

# Modified function to filter by keyword "life"
def scrape_life_quotes(soup, quotes):
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        # Check if "life" is present in the quote text
        if "life" in text.lower():
            author = quote.find('small', class_='author').text
            tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))
            quotes.append({'Text': text, 'Author': author, 'Tags': tags})

def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(response.text, 'html.parser')
        scrape_life_quotes(soup, quotes)
        
        next_page = soup.find('li', class_='next')
        url = 'https://quotes.toscrape.com' + next_page.find('a')['href'] if next_page else None

quotes = []
scrape_all_pages('https://quotes.toscrape.com')

# Save to "life_quotes.csv"
with open('life_quotes.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(quotes)

print("Quotes containing 'life' have been saved to life_quotes.csv")
```
